```python
import pandas as pd
# данные взяты с https://www.kaggle.com/datasets/mdwaquarazam/stock-price-history-top-10-companies?resource=download
df = pd.read_csv('./Libraries/kaggle_data_stock_prices/ZAll_Combine_Stock_Histry.csv')
df.head()
```

```python
df['Year'] = df['Date'].apply(lambda x: x.split('-')[-1])
df['Month'] = df['Date'].apply(lambda x: x.split('-')[-2])


df.head()
```

```python
from plotly import express as px

px.box(df, x = 'Year', y = 'Open',title='Year vs Open price: boxplot',height = 700)
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

plt.figure(figsize = (20,10))

sns.boxplot(data = df, x = 'Year', y = 'Open')
plt.title('Year vs Open price: boxplot');
```

```python
sns.lineplot(data = df, x = 'Month', y = 'Open', hue = 'Year');
```

```python
df[['Month','Year']] = df[['Month','Year']].astype('int')
sns.heatmap(df.corr(),annot=True);
```

```python
sns.catplot(data=df,y='Volume',x='Year')
```

```python
sns.pairplot(data=df, markers = True, )
```

```python
plt.bar(x = df['Year'],height = df['Volume']);
plt.title('Volume')
plt.xlabel('Year')
plt.ylabel('Volume')
plt.yscale('log')
plt.ylim(10**8,10**9);
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(data = df, x = df['Year'],y = df.Open,hue='Company');
plt.title('Open level dynamics over years');
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(data = df, x = df['Year'],y = df.Open,hue='Company');
plt.title('Open level dynamics over years');
```
