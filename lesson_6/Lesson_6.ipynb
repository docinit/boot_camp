```python
from sklearn.model_selection import GridSearchCV as GSCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split as tts

# файл должен быть загружен в каталог с данным ноутбуком;
# источник: https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud?select=creditcard.csv
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
X = df[df.columns.drop('Class')]
```

```python
X
```

```python
y = df.Class
type(y)
```

```python
x_train,x_test,y_train,y_test = tts(X,y,test_size=0.3, random_state=100, stratify=y)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
parameters = [{'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
GridSearchCV = GSCV(estimator=RandomForestClassifier(random_state=100),
                    param_grid=parameters,
                    scoring='roc_auc',
                    cv=3
                    )
GridSearchCV.fit(x_train,y_train)
```

```python
best_params = GridSearchCV.best_params_
print(best_params)
y_train_hat = GridSearchCV.predict_proba(x_train)
y_hat = GridSearchCV.predict_proba(x_test)
y_train_pred_proba = y_train_hat[:,1]
y_pred_proba = y_hat[:,1]
```

```python
from sklearn.metrics import roc_auc_score

print('На тестовых данных:')
print(roc_auc_score(y_test,y_pred_proba))
print('\nНа тренировочных данных:')
print(roc_auc_score(y_train,y_train_pred_proba))
```
